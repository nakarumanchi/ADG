```python
import pandas as pd
```

```python
df = pd.read_csv('test_clean.csv')
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder()
```

```python
for column in df.columns[1:]:
    df[column]=le.fit_transform(df[column])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import numpy as np
```

```python
onehot = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
train = pd.DataFrame(
    onehot.fit_transform(df[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])\
    .toarray(),
    columns=['Female','Male','Not_Married','Married','No_Dependent','1_Dependent','2_Dependents','3+_Dependents','Graduate','Not_Graduate','Not_Self_Employed','Self_Employed','Rural','Semiurban',
             'Urban'])
```

```python
train.head()
```

```python
train.tail()
```

```python
train['Loan_ID'] = df.Loan_ID
```

```python
df.info()
```

```python
train.info()
```

```python
final = pd.merge(train,df[['Loan_ID','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']],on='Loan_ID')
```

```python
final = final.drop('Loan_ID',axis=1)
```
